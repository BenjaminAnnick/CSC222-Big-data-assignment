```python
# Basic imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load the video game sales dataset
df = pd.read_csv('vgsales.csv')

# Display the first 5 rows to check
df.head()
```

```python
#Inspecting the data
df.info()
df.describe()
```

```python
#Checking for missing values
df.isnull().sum()
```

```python
# Handle missing values
df['Year'].fillna(df['Year'].median(), inplace=True)#Replaces missing values with median Year
df['Publisher'].fillna('Unknown', inplace=True)#Replaces missing Publisher with Unknown

# Drop irrelevant column
df.drop(columns=['Rank'], inplace=True)#Drops the Rank column

# Convert Year to integer
df['Year'] = df['Year'].astype(int)

# Confirm cleaning
print(df.isnull().sum())  # Checking to see if missing values still exist
print(df.dtypes)          # Checking the final data types
```

```python
# Quick overview
print(df.info())
print(df.describe())

# Top Platforms
print("Top Platforms:")
print(df['Platform'].value_counts())

# Top Genres
print("\nTop Genres:")
print(df['Genre'].value_counts())

# Top Publishers
print("\nTop Publishers:")
print(df['Publisher'].value_counts().head(10))

# Best-Selling Games
print("\nTop 10 Best-Selling Games Globally:")
print(df[['Name', 'Global_Sales']].sort_values(by='Global_Sales', ascending=False).head(10))

# Global Sales by Platform
print("\nTotal Global Sales by Platform:")
print(df.groupby('Platform')['Global_Sales'].sum().sort_values(ascending=False).head(10))

# Global Sales by Genre
print("\nTotal Global Sales by Genre:")
print(df.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False))


```

```python
# Top 10 platforms by count
top_platforms = df['Platform'].value_counts().head(10)

plt.figure(figsize=(10,6))
plt.bar(top_platforms.index, top_platforms.values)
plt.title('Top 10 Platforms by Number of Games')
plt.xlabel('Platform')
plt.ylabel('Number of Games')
plt.grid(True)
plt.show()
```

```python
# Genre distribution
genre_counts = df['Genre'].value_counts()

plt.figure(figsize=(8,8))
plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Market Share by Genre')
plt.axis('equal')  # Make it a circle
plt.show()
```

```python
sales_by_year = df.groupby('Year')['Global_Sales'].sum()

plt.figure(figsize=(10,6))
plt.plot(sales_by_year.index, sales_by_year.values, marker='o')
plt.title('Global Video Game Sales Trend Over Years')
plt.xlabel('Year')
plt.ylabel('Global Sales (Millions)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['NA_Sales'], df['EU_Sales'], alpha=0.5)
plt.title('NA Sales vs EU Sales')
plt.xlabel('North America Sales (Millions)')
plt.ylabel('Europe Sales (Millions)')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(10,6))
sns.countplot(data=df, y='Platform', order=df['Platform'].value_counts().index[:10])
plt.title('Top 10 Platforms by Number of Games (Seaborn)')
plt.xlabel('Number of Games')
plt.ylabel('Platform')
plt.show()
```

```python
plt.figure(figsize=(14,8))
sns.boxplot(x='Genre', y='Global_Sales', data=df)
plt.title('Global Sales Distribution by Genre')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sales_corr = df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']].corr()

sns.heatmap(sales_corr, annot=True, cmap='coolwarm')
plt.title('Correlation Between Regional Sales')
plt.show()
```
